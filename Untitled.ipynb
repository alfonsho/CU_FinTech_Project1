Say you've got 3 assets, generate matrix A in nxm such that the rows are timelines, the columns are asset

- n : timeline return 
- m : asset

the first matrix is the return for n timelines for the first period. Generate P such matrices for P periods. Chain multiply those matrices for the complete simulation. 

Each column in the matrix R is generated from its own distribution. 

Finally say you've got 3 different portfolios. Generate an ixj matrix W such that 

- i :  weight for asset
- j :  risk. 



```python
def return_matrix():

    a = np.random.normal(0.98, 0.02, 255*5)
    b = np.random.normal(1.02, 0.04, 255*5)
    c = np.random.normal(1.024, 0.1, 255*5)

    return np.matrix([a,b,c]).T

return_matrix()
```

```python
def P():
    p1 = np.array([1/3, 1/3, 1/3])
    p2 = np.array([2/6, 1/6, 3/6])
    p3 = np.array([1/5, 2/5, 2/5])

    return  np.matrix([p1, p2, p3]).T
P = P()
```

```python

for i in range(3000):
    r = return_matrix()
    r = r @ P
r
```
